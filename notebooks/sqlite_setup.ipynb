```python
from google.colab import drive
drive.mount('/content/drive')


import os
import sqlite3

base_path = '/content/drive/MyDrive/IMDB_Data/filtered/'
db_path = '/content/drive/MyDrive/IMDB_Data/imdb_filtered.db'

csv_files = {
    'movies': os.path.join(base_path, 'merged_movies.csv'),
    'principals': os.path.join(base_path, 'principals_filtered.csv'),
    'names': os.path.join(base_path, 'names_filtered.csv')
}


import pandas as pd

conn = sqlite3.connect(db_path)

for table_name, path in csv_files.items():
    print(f"ðŸ”„ Loading {table_name} from {path}")
    df = pd.read_csv(path)
    df.to_sql(table_name, conn, if_exists='replace', index=False)

print("âœ… All filtered CSVs loaded into SQLite.")


def show_schema(table):
    return pd.read_sql(f"PRAGMA table_info({table});", conn)

show_schema('movies')



pd.read_sql("SELECT * FROM movies LIMIT 5;", conn)
```
