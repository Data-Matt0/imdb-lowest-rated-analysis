```python
from google.colab import drive
drive.mount('/content/drive')


import os

base_path = '/content/drive/MyDrive/IMDB_Data/'
filtered_path = os.path.join(base_path, 'filtered')
os.makedirs(filtered_path, exist_ok=True)


import pandas as pd

basics_path = os.path.join(base_path, 'title.basics.tsv')
output_basics = os.path.join(filtered_path, 'filtered_basics.csv')

chunksize = 100000
filtered_titles = []

for chunk in pd.read_csv(basics_path, sep='\t', na_values='\\N', chunksize=chunksize):
    filtered = chunk[(chunk['titleType'] == 'movie') & (chunk['startYear'].astype(float) > 1950)]
    filtered_titles.append(filtered)

df_titles = pd.concat(filtered_titles)
df_titles.to_csv(output_basics, index=False)
print(f"✅ Filtered titles saved to {output_basics}")



ratings_path = os.path.join(base_path, 'title.ratings.tsv')
output_ratings = os.path.join(filtered_path, 'filtered_ratings.csv')

ratings_df = pd.read_csv(ratings_path, sep='\t', na_values='\\N')
ratings_filtered = ratings_df[ratings_df['numVotes'] >= 1000]
ratings_filtered.to_csv(output_ratings, index=False)
print(f"✅ Filtered ratings saved to {output_ratings}")



merged = pd.merge(df_titles, ratings_filtered, on='tconst')
merged_output = os.path.join(filtered_path, 'merged_movies.csv')
merged.to_csv(merged_output, index=False)
print(f"✅ Merged dataset saved to {merged_output}")


merged[['primaryTitle', 'startYear', 'averageRating', 'numVotes']].head(10)
```
